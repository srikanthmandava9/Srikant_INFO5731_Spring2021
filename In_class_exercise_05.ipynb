# Extracting 

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

customer_name = [] # Creating a list that stores names of Customers who provided the ratings and reviews
review_heading = [] # Creating a list that stores the Main Heading Review provided
actual_review = [] # Creating a list that stores the Actual Review provided
number_rating_provided = [] #Creating a list that stores Number Ratings provided

for page_number in range(55):
  URL = "https://www.flipkart.com/samsung-galaxy-a50-blue-64-gb/product-reviews/itmfe4csw5hucdnn?pid=MOBFE4CSF3G2TTUV&page=" + str(page_number) # Creating link based on the page number
  page = requests.get(URL) # Accessing the URL
  soup = BeautifulSoup(page.text, 'html.parser')
  name_of_cust = soup.find_all(class_='_2sc7ZR _2V5EHH') # Fetching the Name of the Customer using Class Name i.e. obtained by using Inspect element on webpage
  heading_rev = soup.find_all(class_='_2-N8zT') # Fetching the Main Heading Review provided using Class Name i.e. obtained by using Inspect on webpage
  actual_rev = soup.find_all(class_='t-ZTKy') # Fetching the Actual Review provided using Class Name i.e. obtained by using Inspect on webpage
  number_rating = soup.find_all(class_='_3LWZlK _1BLPMq') # Fetching the Number Rating provided using Class Name i.e. obtained by using Inspect on webpage
  
  for name, main, actual, rate in zip(name_of_cust, heading_rev, actual_rev, number_rating) : # Iterating through the lists
      customer_name.append(name.text) # Appending name to the empty customer_name list
      review_heading.append(main.text) # Appending heading reviews to the empty review_heading list
      actual_review.append(actual.text) # Appending actual reviews to the empty actual_review list
      number_rating_provided.append(rate.text) # Appending rating to the empty number_rating_provided list

data_frame = pd.DataFrame(list(zip(customer_name, review_heading, actual_review, number_rating_provided)), columns =['Name of the Customer', 'Heading Review', 'Actual Review', 'Customer Rating'])  # Creating Dataframe to create a .CSV file
print("Data frame Length is {0}".format(len(data_frame)))
data_frame # Displaying the contents of the Data frame
```

# PreProcessing

# Lowercase

```python
data_frame['After Preprocessing'] = data_frame['Actual Review'].apply(lambda A: " ".join(A.lower() for A in A.split()))
data_frame
```

# Removing Punctuation

```python
data_frame['After Preprocessing'] = data_frame['Actual Review'].apply(lambda A: " ".join(A.lower() for A in A.split()))
data_frame
data_frame['After Preprocessing'] = data_frame['After Preprocessing'].str.replace('[^\w\s]','')
data_frame
```


# Removing Numbers

```python
data_frame['After Preprocessing'] = data_frame['Actual Review'].apply(lambda A: " ".join(A.lower() for A in A.split()))
data_frame
data_frame['After Preprocessing'] = data_frame['After Preprocessing'].apply(lambda A: ''.join([i for i in A if not i.isdigit()]))
data_frame
```

# Removing special chars

```python
import re
data_frame['After Preprocessing'] = data_frame['Actual Review'].apply(lambda A: " ".join(A.lower() for A in A.split()))
data_frame
data_frame['After Preprocessing'] = data_frame['After Preprocessing'].apply(lambda x: ''.join(re.sub(r"[^a-zA-Z0-9]+", ' ', charctr) for charctr in x ))
data_frame
```

# Download nltk

# Remove stop words

```python
import nltk
from nltk.corpus import stopwords
stop = stopwords.words('english')
data_frame['After Preprocessing'] = data_frame['After Preprocessing'].apply(lambda A: " ".join(A for A in A.split() if A not in stop))
data_frame
```

# TextBold

```python
from textblob import TextBlob
data_frame['After Preprocessing'] = data_frame['After Preprocessing'].apply(lambda A: str(TextBlob(A).correct()))
data_frame
```


# Streaming

```python
import nltk
from nltk.stem import PorterStemmer
stream = PorterStemmer()
data_frame['After Preprocessing'].apply(lambda x: " ".join([stream.stem(word) for word in x.split()]))
```

# Lemmatization

```python
from textblob import Word
import nltk
nltk.download('wordnet')

data_frame['After Preprocessing'] = data_frame['After Preprocessing'].apply(lambda x: " ".join([Word(word).lemmatize() for word in x.split()]))
data_frame
```

### Speech Tagging and Features

```python
from nltk.tokenize import word_tokenize
pos = []
for sentence in data_frame['After Preprocessing']:
  data = word_tokenize(sentence)
  pos.append(nltk.pos_tag(data))
pos
```

```python
Adjective = []
Adverb = []
CordinatingConjunction = []
SubordinatingConjuction = []
Interjection = []
Noun = []
Verb = []
PersonalPronoun = []
predeterminer = []
Determiner = []
```

```python
for value in pos:
  AdjectiveCount = 0
  AdverbCount = 0
  CordinatingConjunctionCount = 0
  SubordinatingConjuctionCount = 0
  InterjectionCount = 0
  NounCount = 0
  VerbCount = 0
  PersonalPronounCount = 0
  predeterminerCount = 0
  DeterminerCount = 0
  for word,tag in value:
    if tag == 'JJ':
      AdjectiveCount = AdjectiveCount + 1
    elif tag == 'RB':
      AdverbCount = AdverbCount + 1
    elif tag == 'CC':
      CordinatingConjunctionCount = CordinatingConjunctionCount + 1
    elif tag == 'UH':
      InterjectionCount = InterjectionCount + 1
    elif tag == 'NN':
      NounCount = NounCount + 1
    elif tag == 'VR':
      VerbCount = VerbCount + 1
    elif tag == 'PRP':
      PersonalPronounCount = PersonalPronounCount + 1
    elif tag == 'PDT':
      predeterminerCount = predeterminerCount + 1
    elif tag == 'DT':
      DeterminerCount = DeterminerCount + 1
    elif tag == 'IN':
      SubordinatingConjuctionCount = SubordinatingConjuctionCount + 1
  Adjective.append(AdjectiveCount)
  Adverb.append(AdverbCount)
  CordinatingConjunction.append(CordinatingConjunctionCount)
  Interjection.append(InterjectionCount)
  Noun.append(NounCount)
  Verb.append(VerbCount)
  PersonalPronoun.append(PersonalPronounCount)
  predeterminer.append(predeterminerCount)
  Determiner.append(DeterminerCount)
  SubordinatingConjuction.append(SubordinatingConjuctionCount)
```

```python
data_frame['Number of Adjectives'] = Adjective
data_frame['Number of Adverbs'] = Adverb
data_frame['Number of Cordinating Conjunctions'] = CordinatingConjunction
data_frame['Number of Interjections'] = Interjection
data_frame['Number of Nouns'] = Noun
data_frame['Number of Verbs'] = Verb
data_frame['Number of Personal Pronouns'] = PersonalPronoun
data_frame['Number of Predeterminers'] = predeterminer
data_frame['Number of Determiners'] = Determiner
data_frame['Number of Subordinating Conjuctions'] = SubordinatingConjuction
data_frame
```

# Right Branching Nodes

```python
RightBranchingNodes = []
import spacy
nlp = spacy.load("en_core_web_sm")
for sentence in data_frame['After Preprocessing']:
  doc = nlp(sentence)
  try:
    RightBranchingNodes.append(doc[0].n_rights)
  except:
    RightBranchingNodes.append('No')
data_frame['Number of Right Branching Nodes'] = RightBranchingNodes
data_frame
```
